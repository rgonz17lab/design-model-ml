```python
import pandas as pd
dataset = pd.read_csv('../dataset/train.csv')
dataset
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(15, 8))
plt.scatter(dataset["GrLivArea"], dataset["SalePrice"], s=50, label='datos')
plt.legend(loc='upper left', fontsize=20)
plt.xlabel('GrLivArea', fontsize=20)
plt.ylabel('SalePrice', fontsize=20)
```

```python
import numpy as np
x = dataset["GrLivArea"].values
y = dataset["SalePrice"].values

mean_x = np.mean(x)
mean_y = np.mean(y)

n = len(x)

numer = 0
denom = 0

for i in range(n):
    numer += (x[i] - mean_x) * (y[i] - mean_y)
    denom += (x[i] - mean_x) ** 2

m = numer / denom
c = mean_y - (m * mean_x)     
print(f'Coefficients:  m: {m}   and c {c}')
```

```python
dataset['y_pred'] = m*x + c
fig = plt.figure(figsize=(12, 6))
plt.scatter(x,y, s=50, label='datos')
plt.plot(x, dataset['y_pred'], color="red", lw=4, label='adjustment')
plt.legend(loc='upper left', fontsize=20)
plt.xlabel('GrLivArea', fontsize=20)
plt.ylabel('SalePrice', fontsize=20)
```
